```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv('./Data sets/DataCoSupplyChainDataset.csv',  encoding_errors='ignore')
df2 = pd.read_csv('./Data sets/DescriptionDataCoSupplyChain.csv',  encoding_errors='ignore')
df3 = pd.read_csv("./Data sets/tokenized_access_logs.csv",  encoding_errors='ignore')


print(df1.head(), df2.head(), df3.head())
```

```python
print(df1.columns)
```

```python
df1['order date (DateOrders)'] = pd.to_datetime(df1['order date (DateOrders)'], errors='coerce')
df1['shipping date (DateOrders)'] = pd.to_datetime(df1['shipping date (DateOrders)'], errors='coerce')
```

```python
df1.fillna({
    'Sales': df1['Sales'].median(),
    'Product Price': df1['Product Price'].median()
}, inplace=True)
```

```python
df1.drop_duplicates(inplace=True)
```

```python
df1['Shipping Delay'] = (df1['shipping date (DateOrders)'] - df1['order date (DateOrders)']).dt.days
df1['Profit Margin'] = df1['Order Item Profit Ratio'] * df1['Product Price']
```

```python
df1 = pd.get_dummies(df1, columns=['Category Name', 'Market', 'Order Region', 'Order Status', 'Product Status', 'Shipping Mode'])
```

```python
columns_to_keep = [
    'Order Id', 'Order Item Id', 'Order Item Product Price', 'Order Item Quantity',
    'Order Item Total', 'Order Profit Per Order',
    'Days for shipping (real)', 'Days for shipment (scheduled)',
    'Shipping Mode_First Class', 'Shipping Mode_Same Day', 'Shipping Mode_Second Class', 'Shipping Mode_Standard Class',
    'Delivery Status', 'Late_delivery_risk',
    'Product Name', 'Product Price', 'Product Category Id',
    'Customer City', 'Customer Country', 'Customer State', 'Customer Zipcode',
    'order date (DateOrders)', 'shipping date (DateOrders)',
    'Order Status_CANCELED', 'Order Status_CLOSED', 'Order Status_COMPLETE', 'Order Status_ON_HOLD',
    'Order Status_PAYMENT_REVIEW', 'Order Status_PENDING', 'Order Status_PENDING_PAYMENT', 'Order Status_PROCESSING',
    'Order Status_SUSPECTED_FRAUD'
]

df_cleaned = df1[columns_to_keep]
```

```python
from sqlalchemy import create_engine
from sqlalchemy.exc import SQLAlchemyError


user = 'postgres'
password = 'asad123'
host = 'localhost'  # or the IP address of your PostgreSQL server
port = '5432'       # default port for PostgreSQL
database = 'Supply Change project'

# Create a connection string
connection_string = f'postgresql://{user}:{password}@{host}:{port}/{database}'

# Create an SQLAlchemy engine
engine = create_engine(connection_string)



try:
    # Attempt to connect to the database
    with engine.connect() as connection:
        print("Connection established successfully!")
        print(f"Connection String: {connection_string}")

except SQLAlchemyError as e:
    print("Failed to connect to the database.")
    print(f"Error: {e}")
```
